```python
str = 'cold'
# enumerate()
list_enumerate = list(enumerate(str))
print('list(enumerate(str) = '
, list_enumerate)
#character count
print('len(str) = '
, len(str))

문자열_1 = "가나다 라마바"
enumerate(문자열_1) #결과물을 하나의 덩어리로 만들어서 가지고있는데 나에게는 덩어리의 주소를 알려주는것, 덩어리를 바로 못가져오기대문에
#그 덩어리를 내가 표현할 수 있는 형태로 바꾸는것이 list 이고 그렇게 해야 값이 표출된다.
print(enumerate(문자열_1)) #원래는 정의된 무언가를 알려줘야하는데 그렇게 나오지 않는 이유 하나의 객체로만들어서 표현해주기 때문.
print(list(enumerate(문자열_1))) #문자열의 index와 값이 표현이 됨 //
#enumerate 문자열이 아니고 다른 값이어도 된다.
```
