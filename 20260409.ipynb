```python
#상속 사용하기

1. 계산기
- 리스트[숫자,문자존재 ,예외처리필요]를 입력을 받고
- +하게되면 리스트값을 다 더하고
- -하게되면 다 빼고 ..*,/

-평균값 ,최대값, 최소값

2.전화번호
개인
- 이름 이메일 전화번호 직장명
- 사용자가 개인정보를 알려주면 저장하고
- 필요하면 불러와서 정보를 주기
- 검색을 통해 특정 정보를 확인하기
```

```python
import random

num_list = []
temp_num = random.randint(5, 10)

try:
    for i in range(temp_num):
        enter = int(input("Enter an integer: "))
        num_list.append(enter)

except ValueError:
    print("정수를 입력해야 합니다!") ## 여기서 리스트 생성 끝

question = int(input("연산을 하려면 1 , 평균, 최대, 최소값을 보기위해선 2 를 눌러주세요."))
if question == 1:
    op = input("연산 기호를 입력하세요 +, -, *, / : ")
    answer = 0

    if op == "+":
        for i in num_list:
            answer += i

    elif op == "-":
        for i in range(len(num_list)):
            if i == 0 :
                answer = num_list[i]
            else :
                answer -= num_list[i]

    elif op == "*":
        answer = 1
        for i in num_list:
            answer *= i

    elif op == "/":
        try:
            if num_list[0] != 0:
                answer = num_list[0]
                for i in num_list[1:]:
                    answer = answer/i

                    if i == 0:
                        raise ZeroDivisionError
            else :
                raise ZeroDivisionError
        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다!")

print(f"입력한 '{op}' 연산 결과 : {answer}")

if question == 2:
    for i in range(len(num_list)):
        answer += num_list[i]
    answer = answer/len(num_list)
    print(f"이 리스트의 평균 값은 '{answer}' 입니다")

    for i in range(len(num_list)):
        for j in range(len(num_list)-1):
            if num_list[j] > num_list[j+1]:
                num_list[j], num_list[j+1] = num_list[j+1], num_list[j]
        max_num = num_list[-1]
        min_num = num_list[0]
    print(f"이 리스트에서의 최댓값은 '{max_num}' 입니다")
    print(f"이 리스트에서의 최솟값은 '{min_num}' 입니다")

else :
    print("지원하지 않는 번호입니다!")
```
