```python
# 문제 : 소수 판별.

# 데이터 가져오기 : 판별 할 숫자를 선언하고 변수에 저장.
# 데이터 가져오기 : 숫자를 나누는 연산을 할 때, 값을 받을 변수를 선언.

# 데이터 연산 : 숫자를 1부터 나누어 보며,
#               나머지가 없이 정확히 나누어 떨어지는지 확인

# 데이터 저장 : 나머지가 생기지 않는 숫자를 새로운 변수에 저장

# 데이터 연산 2 : 숫자를 1씩 늘려가며 변수값까지 반복작업
# 데이터 연산 3 : 반복이 끝났을 때, 저장된 값이 2개 일 때 소수로 판별. > 변수를 list로 선언
#                 정수 type 로 선언을 했을 시, 선언한 숫자에 +1 한 값과 같으면 소수

```

```python
import random as rd

ran_num = rd.randint(-1,20)
boolean = True

if ran_num < 2:
    print('음수, 0, 1은 확인 할 수 없습니다.')
else :
    for i in range(2,ran_num):
        if ran_num%i == 0:
            boolean = False
            break

    if boolean == True :
        print(f'소수 : {ran_num}')
    else:
        print(f'합성수 : {ran_num}')

#작업횟수 ran_num - 2
```

```python
숫자=int(input("숫자 : "))
반환 = None

if 숫자 == 2:
    반환 = True
elif 숫자 > 2:
    for k in range(2, 숫자-1, 1):
        나머지 = 숫자%k
        if 나머지 == 0:
            반환 = False
            break
    if 반환 == None:
        반환 = True

else :
    반환 = False
```

```python
'''# 문제 비밀번호 검증하는 코드
# 1) 길이가 8자 이상.
# 2) 숫자, 문자, 특수문자 조합(!, #, @)
# 3) 대문자가 1개 이상 포함. 제대로 된 비밀번호인지 check 하는 것, 위배약속을 알려줌.

1.비밀번호를 입력받을 변수
- 길이가 8개 미만일 때, 길이가 부족하므로 다시 입력하는 print

2.길이를 만족했을 때, 각각의 비밀번호에 숫자,문자,특수문자가 전부 들어갔는지 확인.
- 만족하지 못하면 (문자,숫자,특수문자)가 없습니다 출력. 다시 입력하게 유도
- 하나씩 꺼냈을때, 문자, 숫자, 특수문자 인지 확인을 한다.

3.대문자 한개 이상 (ord 활용)'''

password = input(("8글자 이상의 비밀번호(대문자 1개 포함, 숫자, 문자, 특수문자 (!, #, @)조합)를 입력하세요. : "))
password = password.replace(" ", "")

if len(password) > 7:
    num_check = False
    text_check = False
    upper_check = False
    spec_check = False

    for check in password:
        if ord('0') <= ord(check) <= ord('9'):
            num_check = True
        elif ord('a') <= ord(check) <= ord('z'):
            text_check = True
        elif ord('A') <= ord(check) <= ord('Z'):
            text_check = True
            upper_chec = True
        elif check in "!@#":
            spec_check = True

    if not num_check:
        print("비밀번호에 숫자를 입력해주세요.")
    if not text_check:
        print("비밀번호에 문자를 입력해주세요.")
    if not upper_check:
        print("비밀번호에 대문자를 입력해주세요.")
    if not spec_check:
        print("비밀번호에 특수문자(!, @, #)를 입력해주세요.")

    if num_check and text_check and upper_check and spec_check:
        print("사용 가능한 비밀번호입니다!")

else :
    print("글자 수를 최소 8글자 이상 적어주세요.")
```
